# Medidas de dispersin

Medidas de dispersin
- Rango: abarca desde el mnimo hasta el mximo en el conjunto de datos (todos los datos)
- Rango intercuartil (IQR): 4 subdivisiones homogneas de los datos
- Desviacin estndar



El diagrama de caja es la visualizacin para representar simplificadamente la dispersin de los datos en referencia a la mediana

continuamos con el dataset https://www.kaggle.com/lepchenkov/usedcarscatalog

## Desviacin estndar

Es la medida de dispersin de datos ms utilizada. Matemticamente es la raz de la media del error cuadrtico. Cuando se est trabajando con muestras en vez de la poblacin, se hace una correccin haciendo que se divida para n-1.

Si los datos siguen una distribucin normal, si se considera todos los datos que estn en un rango de promedio  3*desviacin estndar se estara abarcando el 99.72% de los datos de la distribucin. Los puntos que se salen de eso no corresponden con el patrn y se conocen como outliers y a veces se los descarga. En otras palabras, si los datos estn ms all de 3*std, se descartan.

Si siguen una distribucin asimtrica, no aplica lo anterior, si no que se crean funciones para determinar qu datos son relevantes. Pero generalmente solo se usa para desviacin estndar para distribuciones simtricas.

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('cars.csv')
```

```python
# Desviacin estandar
df['price_usd'].std()
```

```python
# Rango = valor max - valor min
rango = df['price_usd'].max() - df['price_usd'].min()
rango
```

```python
# Quartiles
median = df['price_usd'].median()
Q1 = df['price_usd'].quantile(q=0.25)
Q3 = df['price_usd'].quantile(q=0.75)
min_val = df['price_usd'].quantile(q=0)
max_val = df['price_usd'].quantile(q=1.0)
print(min_val, Q1, median, Q3, max_val)
```

```python
iqr = Q3 - Q1
iqr
```

### Lmites para deteccin de outliers (datos simetricamente distribuidos)

Datos entre $$Q_1 -1.5 \times \text{IQR}$$ y $$Q_3 + 1.5 \times \text{IQR}$$

```python
minlimit = Q1 - 1.5*iqr
maxlimit = Q3 + 1.5*iqr
print('rango para deteccin de outliers: {}, {}'.format(minlimit, maxlimit))
```

El valor es negativo porque se est aplicando una ecuacin de una distribucin simtica a una no simtrica.
