# Análisis de componentes principales
reducción de dimensionalidad por medio de la técnica PCA(análisis de componentes principales)

si hay correlación interesante entre variables, se procede a realizar PCA para eliminar variables que no son necesarias pasarlas en un modelo de marchine learning

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler


iris = sns.load_dataset('iris')

scaler = StandardScaler()
scaled = scaler.fit_transform(
    iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].values
    )

covariance_matrix = np.cov(scaled.T)
covariance_matrix
```

```python
sns.pairplot(iris, hue= 'species' )
```

```python
sns.jointplot(x= iris['petal_length'], y=iris['petal_width'])
sns.jointplot(x = scaled[:, 2], y = scaled[:,3])
```

Descomposicion en vectores y valores propios: https://numpy.org/doc/stable/reference/generated/numpy.linalg.eig.html

```python
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
eigen_values
```

```python
eigen_vectors
```

```python
variance_explained = []
for i in eigen_values:
    variance_explained.append((i/sum(eigen_values))*100)

print(variance_explained)
```

Con el resultado se puede decir que con los dos primeros vectores se acogen el alrededor del 95% de los datos

PCA con scikit : https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html

```python
from sklearn.decomposition import PCA

# NUMERO DE COMPONENETES QUE SE DESEAN TOMAR DE LOS DATOS
pca = PCA(n_components=2)
pca.fit(scaled)
```

```python

pca.explained_variance_ratio_
```

Resultado que representa las dos componentes seleccionadas anteriormente que representan alrededor del 95% de los datos

```python
# transforma los datos originales en los datos reducidos
reduced_scaled = pca.transform(scaled)
```

```python
reduced_scaled
```

```python
iris['pca_1'] = scaled[:,0]
iris['pca_2'] = scaled[:,1]
iris
```

```python
sns.jointplot(x = iris['pca_1'], y = iris['pca_2'], hue = iris['species'])
```
